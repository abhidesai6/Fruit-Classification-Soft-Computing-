```python
import tensorflow.keras.backend as K                  # to access keras function
from tensorflow.keras.models import load_model        # Load model 
from tensorflow.keras.preprocessing import image      # to resize image
import matplotlib.pyplot as plt # plot graphs
import numpy as np   
import os                                            # to use path
import random                                        # to use random mathematic function
import tensorflow as tf   
import tensorflow.keras.backend as K
from tensorflow.keras import regularizers                                 # to regularize overfit model 
from tensorflow.keras.models import Sequential, Model                     # method to train neural network 
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten   # Dense = 
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, ZeroPadding2D, GlobalAveragePooling2D, AveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator   
from tensorflow.keras.callbacks import ModelCheckpoint, CSVLogger
from tensorflow.keras.optimizers import SGD                              # optimizer
from tensorflow.keras.regularizers import l2                              #regularizers
from tensorflow import keras
from tensorflow.keras import models
import cv2
import matplotlib.pyplot as plt



n_classes = 4
hight ,width = 100,100
train_dir = '/content/drive/MyDrive/SoftComputing/Train'

batch_size = 16                       # every 16 batch ,then update

train_datagen = ImageDataGenerator(             
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)            #rescale image . 

test_datagen = ImageDataGenerator(rescale=1. / 255)  # perprocess - rescale-range-flip

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(hight,width),
    batch_size=batch_size,
    class_mode='categorical')

resnet = tf.keras.applications.VGG19(weights = 'imagenet' , include_top = False , input_shape = (100, 100, 3))   #Visual Geometry Group
resnet.trainable = False         # we are freezing 50 layers of Resnet50
x = resnet.output                # output in array 128 x 128 x 256
x = GlobalAveragePooling2D()(x)
x = Dense(128,activation='relu')(x)    # rectified linear activation function 
#x = Dropout(0.2)(x)  
predictions = Dense(n_classes,kernel_regularizer=regularizers.l2(0.005), activation='softmax')(x) 

model = Model(inputs=resnet.input, outputs=predictions)     # concatenating model 
model.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])


history = model.fit_generator(train_generator,
                    steps_per_epoch = 60,   
                    epochs=5
                    )
model.save('Fruit4.hdf5')


plt.title('epochs vs loss')
plt.plot(history.history['loss'])

plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss', 'validation_loss'], loc='best')
plt.show()




```
